```python
res_list = []
for i in range(1,101):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20 , random_state =i)
    dt1 = DecisionTreeClassifier()
    dt1.fit(x_train , y_train)
    s=dt1.score(x_test,y_test)
    res_list.append(s)
    
r=max(res_list)
x=res_list.index(r)
print("random_state:",x+1)
print("score:",r)
```
